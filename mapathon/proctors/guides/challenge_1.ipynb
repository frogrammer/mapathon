inspect data using geopandas

```
# load and plot taxi zone data
taxi_zone_pd = geopandas.read_file('https://s3.amazonaws.com/nyc-tlc/misc/taxi_zones.zip')
ax = taxi_zone_pd.plot()
ax.get_figure().set_size_inches(10, 10)
```

load roads data, show schema

```
roads_pd = geopandas.read_file('https://data.cityofnewyork.us/api/geospatial/svwp-sbcd?method=export&format=GeoJSON')
roads_pd.columns
```

longest roads, project to epsg 2263 to get length in ft

```
roads_pd['road_length'] = roads_pd['geometry'].to_crs(2263).length
ax = roads_pd.sort_values('road_length', ascending=False).head(10000).plot(column='road_length', alpha=0.7)
ax.get_figure().set_size_inches(10, 10)
```

> How many roads are there called 3 AV in Brooklyn? </br>
> How long is 3 AV in Brooklyn? </br>
> Can you show a visualisation of the road within the Brooklyn taxi zone, to verify the result? </br>

```
av3_pd = roads_pd[roads_pd.stname_lab == '3 AV']
av3_zones_pd = geopandas.sjoin(av3_pd.to_crs(2263), taxi_zone_pd, how='inner', op='intersects')
av3_brooklyn_pd = av3_zones_pd[av3_zones_pd.borough == 'Brooklyn']
av3_brooklyn_pd.count()[5]
```

```
av3_brooklyn_pd[['stname_lab', 'borough', 'zone']].head()
```

visualise all parts of road, it looks contiguous

```
brooklyn_pd = taxi_zone_pd[taxi_zone_pd.borough == 'Brooklyn']
ax1 = av3_brooklyn_pd.plot(figsize=(50,50))
brooklyn_pd.plot(ax=ax1, alpha=0.3)
```

sum of lengths

```
# see earlier longest road visualisation for derivation of road_length series
av3_brooklyn_pd['road_length'].sum()
```

too long, approximate using bounds / straight line (minx, miny), (maxx, maxy)

```
from shapely.geometry import LineString, MultiLineString
road_ls = list(av3_brooklyn_pd['geometry'])
road_mls = MultiLineString(road_ls)
road_bounds = road_mls.bounds
LineString([road_bounds[:2], road_bounds[2:]]).length
```

> Can you show the 100 longest roads in Brooklyn, overlaying the Brooklyn taxi zone? </br>

```
road_zones_pd = geopandas.sjoin(roads_pd.to_crs(2263), taxi_zone_pd, how='inner', op='intersects')
brooklyn_pd = road_zones_pd[road_zones_pd.borough == 'Brooklyn']
ax1 = brooklyn_pd.sort_values('road_length', ascending=False)[:100].plot(column='road_length')
ax2 = taxi_zone_pd[taxi_zone_pd.borough == 'Brooklyn'].plot(ax=ax1, alpha=0.3)
ax2.get_figure().set_size_inches(10, 10)
```

> Can you report on any issues found in the road data? <br>
> Can you design an algorithm to address the discovered issues?

- The roads are comprised of many overlapping parts, or parts in different directions for the purposes of navigation.
- To find road lengths, we used a heuristic with bounds. A better method would be to compose a linestring from the parts.
- A method to do this might be to deduce one from buffer of the multilinestring for the road components
